## Assignment-18 Neural Networks

```python
import pandas as pd
import numpy as np
```

```python
column_names = [f'Feature_{i}' for i in range(60)] + ['Target']

data =  pd.read_csv(r'Data\sonardataset.csv',
                    header=None,
                    names=column_names
                   )
data.head()
```

```python
# Assign column names

columns = [f'V{i}' for i in range(1, 61)] + ['target']
data.columns = columns
```

```python
print(data.shape)
```

```python
print(data.info())
```

```python
print(data['target'].value_counts())
```

```python
# missing values

print(data.isnull().sum())
```

##### Encode target variable

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
data['target'] = le.fit_transform(data['target']) # M = 1, R = 0
```

##### Feature scaling (importand for ANN)

```python
print(data.head())
```

```python
# convert all feature columns to numeric

for col in data.columns[:-1]: # exclude target
    data[col] = pd.to_numeric(data[col], errors='coerce')

# drop unnecessary rows
data.dropna(inplace=True)
    
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

##### Scaling

```python
X = data.drop('target', axis=1)
y = data['target']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

##### Train-Test_Split

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,
                                                y,
                                                test_size=0.2,
                                                random_state=42
                                                )
X_train.shape,X_test.shape,y_train.shape,y_test.shape 
```

##### Model Implementation (ANN)

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Dense(64,
                activation='relu',
                input_shape=(X_train.shape[1],),
                kernel_regularizer=l2(0.001)))

model.add(Dropout(0.3))

model.add(Dense(32,
                activation='relu',
                kernel_regularizer=l2(0.001)))

model.add(Dropout(0.3))

model.add(Dense(1, activation='sigmoid'))

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)
```

##### predictions

```python
y_pred = (model.predict(X_test) > 0.5).astype(int)
```

```python
print(y_pred[:5])
```

##### Evalutaion

Metrics

```python
from sklearn.metrics import  f1_score, accuracy_score, precision_score, recall_score
```

```python
print(f'Accuracy_Score:',{accuracy_score(y_test,y_pred)})
print(f'Precision_Score:',{precision_score(y_test,y_pred)})
print(f'f1_Score:',{f1_score(y_test,y_pred)})
print(f'Recall_Score:',{recall_score(y_test,y_pred)})
```

##### Hyperparameter Tuning

Improved Model

```python
model_tuned = Sequential()

model_tuned.add(Dense(64, activation='relu', input_dim=60))
model_tuned.add(Dense(32, activation='relu'))
model_tuned.add(Dense(1, activation='sigmoid'))

model_tuned.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

history_tuned = model_tuned.fit(
    X_train, y_train,
    epochs=100,
    batch_size=8,
    validation_split=0.2,
    verbose=1
)
```

##### Final Evaluation

```python
y_pred_tuned = (model_tuned.predict(X_test) > 0.5).astype(int)

print("Tuned Accuracy:", accuracy_score(y_test, y_pred_tuned))  
```

#### Conclusion

The ANN model successfully classified sonar signals into mines and rocks.

Feature scaling significantly improved model performance.

Hyperparameter tuning increased accuracy and stability.

The tuned model outperformed the baseline model due to deeper architecture and optimized parameters.



#### Overfitting Reduction Techniques

To improve generalization and reduce overfitting, Dropout layers, L2 regularization, and Early Stopping were added to the ANN model.

Dropout helps prevent the network from memorizing training data.
L2 Regularization reduces overly complex weight values.
Early Stopping stops training when validation performance no longer improves.
These techniques help the model achieve better performance on unseen data.
