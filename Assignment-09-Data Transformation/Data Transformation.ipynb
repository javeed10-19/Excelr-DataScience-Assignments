```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r'Data\adult_with_headers (1).csv')
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
X = data.drop('income', axis=1)  # all features
y = data['income']   # target
```

```python
num_cols = X.select_dtypes(include=['int64', 'float64']).columns
cat_cols = X.select_dtypes(include=['object']).columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])
#X[num_cols]
```

```python
# One-hot for small categories
one_hot_cols = [col for col in cat_cols if X[col].nunique() < 5]

# Label encoding for large categories
label_cols = [col for col in cat_cols if X[col].nunique() >= 5]

# One-hot
X = pd.get_dummies(X, columns=one_hot_cols, drop_first=True)

# Label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in label_cols:
    X[col] = le.fit_transform(X[col])
```

```python
le_y = LabelEncoder()
y = le_y.fit_transform(y)
```

```python
data1 = pd.DataFrame(X, columns= data.columns)
```

```python
y = pd.DataFrame(y, columns=['income>50k'])
```

```python
data1 = X.join(y)
```
