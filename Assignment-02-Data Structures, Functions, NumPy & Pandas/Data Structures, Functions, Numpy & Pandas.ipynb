### Section C: Exploring Pandas
#### 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows

```python
import pandas as pd
import numpy as np


data = {
    'Name' : ['Javeed','Tasin','Pavan','venki','Charan',
              'Madhan','Jabeer','Somu','Yashwanth','Saleem'],
    'Age' : [22,21,23,43,54,
             23,24,30,32,19],
    'City' : ['goa','pune','delhi','guntur','chennai',
              'hyd','bglr','gujarat','kerala','pune'],
    'Salary' : [50000, 40000, 60000, np.nan, 80000,
               30000, 25000, 70000, np.nan, 50000]
}
df = pd.DataFrame(data)
df
```

#### Use .info(), .describe()

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

#### Select 'Name' and 'City' columns

```python
df[['Name','City']]
```

#### Drop 'City' column

```python
df.drop('City', axis=1, inplace=True)
df
```

#### Fill any missing values in 'Salary' column with the mean

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df
```

#### Remove any duplicate rows

```python
df.drop_duplicates(subset=['Salary'], inplace=True)
df
```
