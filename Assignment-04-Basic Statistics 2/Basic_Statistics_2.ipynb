#### 1. Load the patient dataset and show summary with info()


```python
patients_df = pd.read_csv(r"Data\Patient_Data.csv")
patients_df.head()
```

```python
billing_df = pd.read_csv('Billing_Data.csv')
billing_df.head()
```

```python
patients_df.info()
```

```python
patients_df.describe()
```

#### 2.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].

```python
billing_columns = ['PatientID','Department', 'Doctor', 'BillAmount']
patients_billing_df = patients_df[billing_columns]
patients_billing_df.head()
```

#### 3. Drop administrative columns like ['ReceptionistID', 'CheckInTime'].


```python
patients_df = patients_df.drop(columns=['ReceptionistID', 'CheckInTime'])
patients_df
```

#### 4.	Use groupby to find total bill amount per department.

```python
patients_df.groupby('Department')['BillAmount'].sum().reset_index()
```

#### 5.	Remove duplicate patient records based on PatientID.

```python
patients_df = patients_df.drop_duplicates(subset='PatientID')
patients_df
```

#### 6.	Fill missing BillAmount values with the mean bill amount.

```python
mean = patients_df['BillAmount'].mean()
mean
```

```python
patients_df.loc[:,'BillAmount'] = patients_df[['BillAmount']].fillna(mean)
patients_df['BillAmount'].isnull().sum()
```

```python
patients_df
```

#### 7.	Merge the billing dataset with patient dataset on PatientID.

```python
merge_df = pd.merge(patients_df,billing_df,  on= 'PatientID')
merge_df.head()
```

#### 8.	Concatenate an additional DataFrame that contains new patients for the current week (row-wise).

```python
new_patients = pd.DataFrame({
    'PatientID':[106,107],
    'Name':['Frank','Grace'],
    'Department':['Neurology','Cardiology'],
    'Doctor':['Dr. John','Dr. Smith'],
    'BillAmount':[6800,7200]
})
```

```python
combined_df = pd.concat([patients_df, new_patients], axis=0).reset_index(drop=True)   # billing data 
combined_df           
```

#### 9.	Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).

```python
insurance_df = pd.DataFrame({
    'InsuranceCovered':[2000,1500,2500,3000,1000],
    'FinalAmount':[3000,3500,5000,3200,4000]
})
```

```python
final_df = pd.concat([combined_df, insurance_df], axis=1)
final_df
```
