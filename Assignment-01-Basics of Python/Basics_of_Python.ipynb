### Exercise 1: Prime Numbers


#### Q. Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.


```python
num = int(input("Enter a number: "))

if num <= 1:
    print(f"{num} is not a prime number.")
else:
    for i in range(2,num):
        if num % i != 0:
            print(f"{num} is not a prime number.")
            
            break
    else:
        print('Prime Number')
```

### Exercise 2: Product of Random Numbers



#### Q. Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.


```python
import random

#Generating two random numbers from 1 to 100
num1 = random.randint(1,100)
num2 = random.randint(1,100)

user_answer = int(input(f'What is {num1} *{num2} ='))
answer = num1 * num2

if answer == user_answer:
    print('Correct Answer!!')
else:
    print(f'Incorrect. The Correct Answer is {answer}')
```

### Exercise 3: Squares of Even/Odd Numbers



#### Q. Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.

```python
# Odd numbers in the range (100,200)
for num in range(100,200):
    if num % 2 != 0:
        square = num ** 2
        print(f'Number: {num} square: {square}')
```

#### Problem Statement (Question): Hospital Billing System
A hospital charges its patients based on the following:

Room charges per day: ₹2,000

Doctor consultation fee (flat): ₹1,500

Lab test charges: Based on the number of tests taken. Each test costs ₹300

Medicine charges: Total cost of medicines provided

Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied

#### Write a Python program to calculate the final bill for a patient who:
Stayed for 4 days

Had 3 lab tests

Had medicine charges of ₹2,400

```python
# Hospital Billing System

room_charges = 2000
docter_consultaion_fee = 1500
lab_test_charges = 300

num1 = int(input('Enter total days of stay:'))
num2 = int(input('Enter total lab tests:'))

total_room_charges = num1 * room_charges
total_lab_tests = num2 * lab_test_charges
medicine_charges = int(input('Enter total medicine charges:'))

print('---------Hospital Blling---------')
print('Room Charges:',total_room_charges)
print('Lab Tests:',total_lab_tests)
print('Medicine Amount:',medicine_charges)
print('Doctor consulation Fees:',docter_consultaion_fee)

total_bill = total_room_charges + total_lab_tests + medicine_charges + docter_consultaion_fee
print('Total Hospital Billing Amount:',total_bill)

# Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied

if total_bill > 10000:
    discount = (total_bill * 10)/100
    final_bill = total_bill-discount
    print('Discount:',discount)
    print('Final bill to pay:',final_bill)
else:
    print('Final bill to pay:',final_bill)
```
